## Pandas

### The read_csv() Method

Read data from `Salaries.csv`. Use `playerID` as the index column, first row as header, and skip the second row.

````python
df = pd.read_csv('../../Homewrok 0 Data/Salaries.csv',
                 index_col='playerID',
                 header=0,
                 skiprows=[1])
df.head()
````

### Indexing and Selecting Data

Select the id of players registered in ATL and HOU whose salary is higher than one million.

````python
result = df[df['teamID'].isin(['ATL', 'HOU']) & df['salary'] > 1000000]
````

### The describe() Method

Calculate the standard deviation, Q1, median, Q3, mean, max, and min of the salary in team ATL.

````python
atl_salary = df[df['teamID'] == 'ATL']['salary']
atl_salary.describe()
````

### The iterrows() Method

Create a Python dictionary whose keys are the dataframe headers and values are lists containing the corresponding column data. Use `iterrows()` to iterate each row.

````python
data_dict = {col : [] for col in df.columns}

for _, row in df.iterrows():
    for col in df.columns:
        data_dict[col].append(row[col])

data_dict
````

### Create Dataframe Using the Constructor

Create a dataframe using `pd.DataFrame` from the dictionary above. Change the headers to "a", "b", "c", ... at creation time.

````python
import string
new_columns = list(string.ascii_lowercase[:len(df.columns)])

df2 = pd.DataFrame(data_dict, columns=df.columns)
df2.columns = new_columns
df2.head()
````

## Numpy

Create a 2-dimensional Python list, then convert it to a Numpy array.

````python
ls_2d = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
arr = np.array(ls_2d)
arr
````

### ndarray Attributes

Play with `ndim`, `shape`, `size`, `dtype`, `itemsize`, and `data`.

````python
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.itemsize)
print(arr.data)
````

### The Slice Operation of ndarray Objects

Understand how slicing works for 1-D and 2-D arrays.

````python
# 1-D
a = np.array([10, 20, 30, 40, 50])
print(a[1:4])      
print(a[::2])      

# 2-D
print(arr[1:])
print(arr[1:, 0:2])
print(arr[:, 1])
print(arr[:, 1:2])
````

### The Calculation of ndarray Objects

Play with `argmin()`, `argmax()`, `min()`, `max()`, `mean()`, `sum()`, `std()`, `dot()`, `square()`, `sqrt()`, `abs()`, `exp()`, `sign()`, `mod()`.

````python
print(np.argmin(arr), np.argmax(arr))
print(arr.min(), arr.max())
print(arr.mean(), arr.sum(), arr.std())
print(np.dot(arr, arr))
print(np.square(arr))
print(np.sqrt(arr))
print(np.abs(arr))
print(np.exp(arr))
print(np.sign(arr - 5))
print(np.mod(arr, 3))
````

## Matplotlib

### (a) Basic Line Plot

Create arrays `x` and `y`, plot y vs x with title, xlabel, ylabel, and grid.

```python
x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])
```

````python
x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])

plt.plot(x, y)
plt.title('Parabola y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
````

### (a continued) Unsorted x

What happens if x is not sorted before plotting? Try with `np.random.shuffle(x)`.

````python
x = np.linspace(-5, 5, num=20)
np.random.shuffle(x)
y = np.array([j ** 2 for j in x])

plt.plot(x, y)
plt.title('Unsorted x — zigzag result')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
````

### (b) Multiple Curves with Styles and Legends

Plot y1, y2, y3 with different styles, add legends and mathematical equation annotations.

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
```

````python
x = np.linspace(-5, 5, num=20)
y1 = x
y2 = x ** 2
y3 = x ** 3

plt.plot(x, y1, 'r--',  label='$y = x$')
plt.plot(x, y2, 'bs-',  label='$y = x^2$')
plt.plot(x, y3, 'g^:',  label='$y = x^3$')
plt.legend()
plt.title('Three curves')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.text(-4, 50, r'$f(x)=x^n$', fontsize=12)
plt.show()
````

### (c) Multiple Curves in One Figure (No subplots)

Plot y1, y2, y3 all in one figure without using subplots.

````python
x = np.linspace(-5, 5, num=20)
y1, y2, y3 = x, x**2, x**3

fig, ax = plt.subplots()   # single axes, no subplot grid
ax.plot(x, y1, label='y1')
ax.plot(x, y2, label='y2')
ax.plot(x, y3, label='y3')
ax.legend()
ax.grid(True)
plt.show()
````

### (d) Subplots

Create multiple subplots (one curve per subplot). Play with figure size, title, font style and size.

````python
x = np.linspace(-5, 5, num=20)
y1, y2, y3 = x, x**2, x**3

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
fig.suptitle('Three Functions', fontsize=16, fontweight='bold')

for ax, y, label in zip(axes, [y1, y2, y3], ['y=x', 'y=x²', 'y=x³']):
    ax.plot(x, y)
    ax.set_title(label, fontsize=13, fontstyle='italic')
    ax.grid(True)

plt.tight_layout()
plt.show()
````

### (e) Axis Limits and Logarithmic Axes

Change x/y axis limits and use logarithmic axes to plot the parabola.

````python
x = np.linspace(-5, 5, num=20)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y)
axes[0].set_xlim(-6, 6)
axes[0].set_ylim(0, 30)
axes[0].set_title('Custom axis limits')
axes[0].grid(True)

# log y axis (y>0 only)
x_pos = np.linspace(0.1, 5, 50)
axes[1].semilogy(x_pos, x_pos ** 2)
axes[1].set_title('Logarithmic y-axis')
axes[1].grid(True)

plt.tight_layout()
plt.show()
````

## Pandas DataFrame.plot and Seaborn

### Pandas DataFrame.plot

#### (a) Scatter Plot — ATL: yearID vs salary

````python
df_atl = df[df['teamID'] == 'ATL']
df_atl.plot.scatter(x='yearID', y='salary', title='ATL: yearID vs Salary')
plt.show()
````

#### (b) Bar Chart — Average salary per team in 1985

````python
df_1985 = df[df['yearID'] == 1985]
df_1985.groupby('teamID')['salary'].mean().plot.bar(
    title='Average Salary per Team (1985)', figsize=(14, 4))
plt.ylabel('Average Salary')
plt.tight_layout()
plt.show()
````

#### (c) Line Chart — Annual average salary change for ATL

````python
df_atl.groupby('yearID')['salary'].mean().plot.line(
    title='ATL Annual Average Salary', marker='o')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
````

### Seaborn

#### (a) Pairplot — ATL with an added random numeric feature

````python
df_atl2 = df_atl[['yearID', 'salary']].copy()
df_atl2['random_feature'] = np.random.randn(len(df_atl2))

sns.pairplot(df_atl2)
plt.suptitle('ATL Pairplot', y=1.02)
plt.show()
````

#### (b) Boxplot — Salary distribution per team in 1985

````python
plt.figure(figsize=(14, 5))
sns.boxplot(data=df_1985, x='teamID', y='salary')
plt.title('Salary Distribution per Team (1985)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

#### (c) Explore lmplot, catplot, relplot, and jointplot

````python
df_atl3 = df_atl.reset_index()

# lmplot: regression line
sns.lmplot(data=df_atl3, x='yearID', y='salary')
plt.title('lmplot')
plt.show()

# catplot: categorical
sns.catplot(data=df_1985, x='teamID', y='salary', kind='box', height=4, aspect=3)
plt.title('catplot')
plt.show()

# relplot: scatter with hue
sns.relplot(data=df_atl3, x='yearID', y='salary', kind='scatter')
plt.title('relplot')
plt.show()

# jointplot: bivariate
sns.jointplot(data=df_atl3, x='yearID', y='salary', kind='reg')
plt.suptitle('jointplot', y=1.02)
plt.show()
````
